```
import pandas as pd

url = "https://raw.githubusercontent.com/IBM/employee-attrition-aif360/master/data/emp_attrition.csv"
df = pd.read_csv(url)

print(df.head())
```

```
df.info()
```

```
df['Attrition'].value_counts()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()
```

```
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title("Age vs Attrition")
plt.show()
```

```
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("OverTime vs Attrition")
plt.show()
```

```
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

Project Title

HR Employee Attrition Analysis

Aim

To analyze employee attrition and identify the factors that influence employees to leave the organization.

Tools Used
Python
Google Colab
Pandas
NumPy
Matplotlib
Seaborn
Dataset Description

The dataset contains employee information such as age, salary, department, overtime, job satisfaction, and attrition status.

Findings
Overtime employees show higher attrition.
Low-income employees are more likely to leave.
Job satisfaction affects attrition.
Younger employees tend to have higher attrition rates.
Conclusion

Employee attrition can be reduced by improving work-life balance, increasing employee satisfaction, and providing better career growth opportunities.
